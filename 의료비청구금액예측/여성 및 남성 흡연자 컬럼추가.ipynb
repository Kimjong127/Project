### 새로운 컬럼 추가

```python
# 여성흡연자 or 비흡연자

data['female_smoker'] = np.where((data['sex'] == 'female') & (data['smoker'] == 'yes'), 1, 0)
data['female_smoker'].value_counts()
```

```python
data.groupby(['female_smoker'])[['charges']].mean()
```

```python
# 남성흡연자 or 비흡연자

data['male_smoker'] = np.where((data['sex'] == 'male') & (data['smoker'] == 'yes'), 1, 0)
data['male_smoker'].value_counts()
```

```python
data.groupby(['male_smoker'])[['charges']].mean()
```

```python
# BMI 비만 or 비만이 아닌 그룹 나누기

data['obese'] = np.where(data['bmi'] >= 30, 1, 0)
```

```python
# 비만 or 비만이 아닌 그룹의 진료비 평균 

data.groupby(['obese'])[['charges']].mean()
```
